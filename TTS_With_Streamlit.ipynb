```python
import streamlit as st
```

```python
st.title("All about Ava.ai")

"""
We are building our most capable Generative ai companion for mental health and mental wellness having
recommendations, suggestions and giving you that human touch whenever u talk to it
"""

st.subheader("Converting text to speech")

text = st.text_input("enter your text here...", value="")

pipe = pipeline("text-to-speech", model="suno/bark-small")
print(pipe)
output = pipe(text)
st.audio(output["audio"], sample_rate=output["sampling_rate"])
```
